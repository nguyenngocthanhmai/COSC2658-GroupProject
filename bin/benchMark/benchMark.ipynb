```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv('benchmark_results.csv')

# Create pivot tables for the heat maps
pivot_search_time = data.pivot("Depth", "NodeCount", "SearchTime")
pivot_init_time = data.pivot("Depth", "NodeCount", "InitializationTime")
pivot_memory_usage = data.pivot("Depth", "NodeCount", "MemoryUsage")
```

```python
# Create a figure with 3 subplots, one for each heat map, arranged vertically
fig, axes = plt.subplots(3, 1, figsize=(10, 24))
# Plot the heat map for Search Time
sns.heatmap(pivot_search_time, annot=True, fmt=".2f", cmap="coolwarm", ax=axes[0])
axes[0].set_title('Heat Map of Search Time by Depth and Node Count')
axes[0].set_xlabel('Node Count')
axes[0].set_ylabel('Depth')

# Plot the heat map for Initialization Time
sns.heatmap(pivot_init_time, annot=True, fmt=".0f", cmap="viridis", ax=axes[1])
axes[1].set_title('Heat Map of Initialization Time by Depth and Node Count')
axes[1].set_xlabel('Node Count')
axes[1].set_ylabel('Depth')

# Plot the heat map for Memory Usage
sns.heatmap(pivot_memory_usage, annot=True, fmt=".0f", cmap="viridis", ax=axes[2])
axes[2].set_title('Heat Map of Memory Usage by Depth and Node Count')
axes[2].set_xlabel('Node Count')
axes[2].set_ylabel('Depth')

# Adjust layout
plt.tight_layout()
plt.show()
```
